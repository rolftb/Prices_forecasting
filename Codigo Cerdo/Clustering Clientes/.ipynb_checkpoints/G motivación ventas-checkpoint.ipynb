````python
# =============================================================================
# Importar paquetes y abrir Data
# =============================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from geopandas import *
from plotnine import *

#para ajustar las fechas
import datetime

#para calcular el skew de un array() o list()
from scipy.stats import skew 

#mode function
import statistics 

col_names=pd.read_csv("Data/(column)Facturac_2017_2020.csv",
                    encoding="utf-8",sep=";",nrows=0).columns
types_dict = {col: str for col in col_names}

Data = pd.read_csv("Data/(column)Facturac_2017_2020.csv",
                   encoding="utf-8",sep=";",dtype=types_dict)




Data=Data[Data['Zona de ventas.1']=='Temuco']
print(Data.describe(include='all'))
````

````python
# =============================================================================
# Ajustar la data
# =============================================================================
def Ajuste_data(data):
    df=data.copy()
    # Funciones para eliminar los putos y cambiar las , por .
    def change_num_point(x):
        x=str(x)
        return x.replace('.', '')
    def change_num_coma(x):
        x=str(x)
        return x.replace(',', '.')
    for i in ['Kilos Venta KG','Venta Neta CLP','Precio Promedio CLP']:
        df[i]=df[i].apply(change_num_point)
        df[i]=df[i].apply(change_num_coma)
        df[i]=df[i].astype('int64')

        #Separacion fecha
    df['Semana']=df['Año natural/Semana'].str.split('.', expand=True)[[0]]
    df['Año']=df['Año natural/Semana'].str.split('.', expand=True)[[1]]

    df['Semana']=df['Semana'].astype('int64')
    df['Año']=df['Año'].astype('int64')
    
    #
    df["Día natural"] = df["Día natural"].apply(lambda x: 
                                                datetime.datetime.strptime(x,
                                                                           '%d.%m.%Y').date())
    return df
````

````python
data=Ajuste_data(Data)
data=data[data['Año']>2018]
````

````python
#id_Cluster='2'
#df_i=DF_ventas[DF_ventas.identified_clusters==2]
def data_vol_clust(df_i,id_Cluster):

    
    
    ano_in=2020
    # mes de inicio
    #int((str(df_i['Día natural'].min()))[5:7])
    #mes_in=df_i['Día natural'].min().month
    mes_in=1

    #---------------------------------------------------------------------------
    # values
    #---------------------------------------------------------------------------

    date_v=list(df_i.groupby(['Mes','Año'],
                             sort=True).count()['Cliente - Local.1'])

    #index
    date_i=list((df_i.groupby(['Mes','Año'],
                              sort=True).sum()['Kilos Venta KG']).index)
    date_i


    #------------------------------
    #
    #------------------------------
    fechas=list()
    for a in range(ano_in,2021):
        if a==ano_in:
            for m in range(mes_in,13):
                t=tuple((str(m),a))
                fechas.append(t)     
        elif(a==2021):
            for m in range(1,3):
                t=tuple((str(m),a))
                fechas.append(t)
        else:
            for m in range(1,13):
                t=tuple((str(m),a))
                fechas.append(t)


    DF_i=((pd.merge(pd.DataFrame({'Date':fechas}),
                    pd.DataFrame({'Date':date_i,'Volumen de venta KG':date_v})
                 , how='left')))



    DF_i['mes']=DF_i['Date'].apply(lambda x : x[0])
    #print(DF_i['Date'])
    V_mean=DF_i.groupby('mes').mean()#['Volumen de venta KG']

    DF_i_R=pd.DataFrame({'Mes':list((V_mean).index),
                         'Cantidad de clientes del cluster':V_mean['Volumen de venta KG'],

                         'Cluster de Clientes':[id_Cluster 
                                                for x in range(len(list((V_mean).index)))]    
    })
    DF_i_R['Mes']=DF_i_R['Mes'].astype('int64')
    DF_i_R=DF_i_R.sort_values(by=['Mes'])
    return(DF_i_R)
````

# Consumo promedio de cada mes

## Promedio base Volumen
\begin{equation}
Precio Promedio= \dfrac{\sum Ingreso}{\sum Volumen} = 
\sum \dfrac{Precio*Volumen}{Volumen total}    
\end{equation}

### Función

````python
col_Clust='identified_clusters'
df_i=DF_ventas[DF_ventas[col_Clust]==2].copy()
df_i=df_i[df_i.Año==2020]
id_Cluster='Cluster 2'

def Proemdio_vol(df_i,id_Cluster):
    
    X=df_i[df_i['Año']==2020].groupby(['Año','Mes']).sum().reset_index()
    X.Mes=X.Mes.astype('int64')
    X=X.sort_values(by=['Mes'])    
    # return x
    if(id_Cluster!='Sin Cluster'):
        Cluster_df_vol_i= Cluster_df_vol[
            Cluster_df_vol['Cluster de Clientes']==str(id_Cluster)] 

        DF_Sim=pd.merge(X[['Año', 'Mes','Kilos Venta KG','Venta Neta CLP']] ,
                 Cluster_df_vol_i[['Año', 'Mes', 'Cantidad de Clientes','Total de Clientes\n registrado hasta la fecha'
                                   ,'Porcentaje de \n los Clientes','Cluster de Clientes']]
                                   , how='left')
        DF_Sim=DF_Sim.rename(columns={'Kilos Venta KG':'Kilos Venta KG Total','Venta Neta CLP':'Venta Neta CLP Total'
                              })
        #DF_Sim['identified_clusters']=[('Cluster '+str(id_Cluster)) for i in range(DF_Sim.shape[0])]
    else:
        Cluster_df_vol_i= Cluster_df_vol[
        Cluster_df_vol['Cluster de Clientes']==id_Cluster] 

        DF_Sim=pd.merge(X[['Año', 'Mes','Kilos Venta KG','Venta Neta CLP']] ,
                 Cluster_df_vol_i[['Año', 'Mes', 'Cantidad de Clientes','Total de Clientes\n registrado hasta la fecha'
                                   ,'Porcentaje de \n los Clientes','Cluster de Clientes']]
                                   , how='left')
        DF_Sim=DF_Sim.rename(columns={'Kilos Venta KG':'Kilos Venta KG Total','Venta Neta CLP':'Venta Neta CLP Total'
                              })
        #DF_Sim['identified_clusters']=[(id_Cluster) for i in range(DF_Sim.shape[0])]
    DF_Sim['Precio promedio CLP ']=DF_Sim['Venta Neta CLP Total']/DF_Sim['Kilos Venta KG Total']
    DF_Sim['Ingreso de \n 1 Cliete']=DF_Sim['Venta Neta CLP Total']/DF_Sim['Cantidad de Clientes']
    DF_Sim['Consumo de \n 1 Cliete']=DF_Sim['Kilos Venta KG Total']/DF_Sim['Cantidad de Clientes']
    DF_Sim['Toneladas de Venta TG Total']=DF_Sim['Kilos Venta KG Total']/1000      
    return(DF_Sim)
Proemdio_vol(df_i,id_Cluster)
````

## PROMEDIO VENTAS
\begin{equation}
\sum\dfrac{Precio}{N° de ventas}
\end{equation}

````python
col_Clust='identified_clusters'
id_Cluster=2
df_i=DF_ventas[DF_ventas[col_Clust]==id_Cluster].copy()
df_i=df_i[df_i.Año==2020]
def Mean_mond(df_i,id_Cluster):
    x=df_i.groupby(['Año','Mes']).mean().reset_index()
    x.Mes=x['Mes'].astype('int64')
    x.sort_values(by=['Mes'])
    return(x)
Mean_mond(df_i,id_Cluster)
````

````python
col_Clust='identified_clusters'
List_Cluster=Cluster.identified_clusters.unique()

id_Cluster='Sin Cluster'#List_Cluster[i]
df_i=df_i[df_i.Año==2020]
df_i=DF_ventas[DF_ventas[col_Clust].isnull()].copy()

#df_i=df_i[df_i.Año==2020]

#función
df_prec=Mean_mond(df_i,id_Cluster)

for i in range(0,5):
    id_Cluster='Cluster ' +str(i)#List_Cluster[i]
    df_i=DF_ventas[DF_ventas[col_Clust]==i].copy()
    df_i=df_i[df_i.Año==2020]
    #función
    df_prec=pd.concat([df_prec,Mean_mond(df_i,id_Cluster)])
df_prec.head(10)
````

````python
df_prec.identified_clusters=df_prec.identified_clusters.fillna(-1)

def cam_etiq(x): 
    S=''
    Segmento={(S+'2'):'raro',
              (S+'1'):'cliente tipo',
              (S+'4'):'Mayorista',
             ( S+'3'):'Oportunista menor',
             ( S+'0'):'Frecuente y Atractivo',
             ( '-1'):'Sin Cluster'
             }
    x=int(x)
    x=str(x)
    return(Segmento[x])

df_prec.identified_clusters=df_prec.identified_clusters.astype('category')


df_prec.identified_clusters=df_prec.identified_clusters.apply(cam_etiq)
````
