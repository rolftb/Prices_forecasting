## Función

```python
df_null=Precio_prom_y_vol.iloc[6853:6856].copy()
def Col_F(x,Col_i):
    mean_i=0
    n=0
    for c in  Col_i:
        if(np.isnan((x[c]))!=True):
            #print('nuemro',x[c])
            mean_i+=x[c]
            #print(mean_i)
            n+=1
            #print(n)
    mean_i=(mean_i/n)
    #print('mean_i',mean_i)
    for c in Col_i:
        if(np.isnan((x[c]))):
            x[c]=mean_i
    return(x)
df_null[Col_P].apply(lambda x: Col_F(x,Col_P), axis=1 )
for i in C_rep:
    print(i)
    df_null[i]=df_null[i].apply(lambda x: Col_F(x,i), axis=1 )
print(df_null.isnull().sum() )
df_null
```

## Columnas con NaN

```python
def Grafico_Nan(S):
    #S='Oficina de Ventas Unifica'
    Col=list(D_Clientes.columns)
    Col=Col[Col.index('Covarianza del lote'):]

    #PorcNaN=[ (D_Clientes[i].isnull().sum()/D_Clientes.shape[0])*100 for i in Col]
    Atributo=list()
    PorcNaN=list()
    S_Val=list()
    for c in Col:
        for s in D_Clientes[S].unique():
            Atributo.append(c)
            PorcNaN.append((D_Clientes[
                D_Clientes[S]==s][c].isnull().sum()/D_Clientes.shape[0])*100)
            S_Val.append(s)
    G=(
        ggplot(pd.DataFrame({'Atributo':Atributo,
                             'Procentaje de NaN':PorcNaN
                             ,S:S_Val
                            })
               , aes(x='Atributo',
                     y='Procentaje de NaN',fill=S))\
    + geom_col()\
    +theme(axis_text_x = element_text(angle=90, hjust=1)))#
    return(G)
Graf=list()
for i in list(Categ_Atrib.columns)[1:]:
    Graf.append(Grafico_Nan(i))
Graf
```

```python
a=X.groupby('identified_clusters').mean()#['Frecuencia mes promedio']
b=X.groupby('identified_clusters').std()#['Frecuencia mes promedio']

def GRaphP(C_i,T_i,Y_i):
    Value=list()
    T_sem=list()
    Indent_cluster=list()
    cov=list()
    for i in list(a.index):
        for j in C_i:
            Value.append(a.iloc[i][j])
            cov.append(b.iloc[i][j]/a.iloc[i][j])
            T_sem.append(j[len(j)-1])#S_Clust[j])
            Indent_cluster.append(i)
    P=pd.DataFrame(
    {
    'Value':Value,
    'Covarianza \n del Cluster':cov,
    'Tipo de Semana':T_sem,
    'Cluster de Clientes':Indent_cluster,
    })
    P['Cluster de Clientes']=P['Cluster de Clientes'].astype('category')
    P['Tipo de Semana']=P['Tipo de Semana'].astype('int64')
    return(P) 
```

```python
def Graph_T_S(P,C_i,T_i,Y_i):
    G=(
            ggplot(P, aes(x='Tipo de Semana',y='Value'
                             ))\
                        + labs(y=Y_i)\
                        + geom_point(aes(color='Cluster de Clientes',
                                         size='Covarianza \n del Cluster'))\
                         + geom_line(aes(color='Cluster de Clientes'))\
                        +ggtitle(T_i)\
                        #+theme(axis_text_x = element_text(angle=90, hjust=1))
                               #,figure_size=(15,8)

            )
    return(G)
```

### Grafico G_T_S

```python
G_T_S=list()
for i in range(len(C_rep)):
    G_T_S.append(Graph_T_S(GRaphP(C_rep[i],T_rep[i],Y_rep[i]),C_rep[i],T_rep[i],Y_rep[i]))
    if i==0:
        dfP=GRaphP(C_rep[i],T_rep[i],Y_rep[i])
        dfP=dfP[dfP['Cluster de Clientes']!=2]
        G_T_S.append(Graph_T_S(dfP,C_rep[i],T_rep[i],Y_rep[i]))
G_T_S   
```
